## 11) Using the first three `mbids`, request the API urls and print the artist's name and their tags

## 12) Using the first ten mbids, print the artist's name and whether they're a rapper

Only print their name ONCE and only print whether they are hip hop or not ONCE.

- *Tip: Rap tags include hip hop, swag, crunk, rap, dirty south, and probably a bunch of other stuff! You can include as many categories as you'd like.*
- *Tip: You can use `2 in [1, 2, 3]` to find out if `2` is in the list of `[1, 2, 3]`.*
- *Tip: Every time you look at a new artist, you can say they are NOT a rapper. And once you find out one of their tags is hip hop or rap, then you can note that they're a rapper. Then once you're done looking at their tags, then you can say HEY this is a rapper, or HEY this is not a rapper.*

````python
mbids = mbids[:10]

rapper_tags = ['hip hop', 'rap', 'swag', 'crunk', 'dirty south']

for mbid in mbids:
    url = f"https://ws.audioscrobbler.com/2.0/?method=artist.getInfo&mbid={mbid}&api_key=87c3d441ee2d00a0fada24c26d5d510b&format=json"
    response = requests.get(url)
    data = response.json()
        
    if 'artist' in data:
            artist_name = data['artist'].get('name', 'Unknown Artist')
            print(f"Artist: {artist_name}")

            is_rapper = False
            if 'tags' in data['artist']:
                for tag in data['artist']['tags']['tag']:
                    if tag['name'].lower() in rapper_tags:
                        is_rapper = True
                        break  

            if is_rapper:
                print("This artist is a rapper.")
            else:
                print("This artist is NOT a rapper.")
    
    print()
````

## 13) What percent of "lil" results are rappers?

````python
rapper_tags = ['hip hop', 'rap', 'swag', 'crunk', 'dirty south']

lil_count = 0
lil_rapper_count = 0

for mbid in mbids:
    url = f"https://ws.audioscrobbler.com/2.0/?method=artist.getInfo&mbid={mbid}&api_key=87c3d441ee2d00a0fada24c26d5d510b&format=json"
    response = requests.get(url)
    data = response.json()
    
    if 'artist' in data:
        artist_name = data['artist'].get('name', 'Unknown Artist')
        
        if 'lil' in artist_name.lower():
            lil_count += 1
            print(f"Artist: {artist_name}")
            
            is_rapper = False
            if 'tags' in data['artist']:
                for tag in data['artist']['tags']['tag']:
                    if tag['name'].lower() in rapper_tags:
                        is_rapper = True
                        break  
            if is_rapper:
                print("This artist is a rapper.")
                lil_rapper_count += 1
            else:
                print("This artist is NOT a rapper.")
    
    print()  
if lil_count > 0:
    rapper_percentage = (lil_rapper_count / lil_count) * 100
    print(f"Percentage of 'Lil' artists who are rappers: {rapper_percentage:.2f}%")
else:
    print("No 'Lil' artists found in the provided mbids.")
````
